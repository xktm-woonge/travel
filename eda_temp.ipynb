```python
month_temp = []
for i in month_day_data.월.unique():
    month = month_day_data[month_day_data['월'] == i]
    month_temp.append({
        '월' : i,
        '최대' : round(month['기온(°C)'].max(), 1),
        '최소' : round(month['기온(°C)'].min(), 1)
     })
```

```python
all_mean_temp = pd.DataFrame(month_temp)
```

```python
for i in range(len(all_mean_temp)):
    all_mean_temp.loc[i, '차이'] = all_mean_temp.loc[i, '최대'] - all_mean_temp.loc[i, '최소'] 
```

```python
plt.figure(figsize=(10,5))
bar = plt.bar(all_mean_temp['월'], all_mean_temp['차이'])
plt.xticks(all_mean_temp['월'])
plt.title('최대 최소 기온 차 (전국 10년 평균)')
plt.xlabel('월')
plt.ylabel('기온')
for rect in bar:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, height, '%.1f' % height, ha='center', va='bottom', size = 12)
```

```python
plt.figure(figsize=(10,5))
max_temp = plt.bar(all_mean_temp['월'],all_mean_temp['최대'], label='최대')
min_temp = plt.bar(all_mean_temp['월'],all_mean_temp['최소'], label='최소')
plt.xticks(all_mean_temp['월'])
plt.ylim([-2, 30]) 
plt.title('최대 최소 기온 차 (10년 평균)')
plt.xlabel('월')
plt.ylabel('기온')
plt.legend(loc='upper right')
for rect in max_temp:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, height, '%.1f' % height, ha='center', va='bottom', size = 12)
for rect in min_temp:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, height, '%.1f' % height, ha='center', va='bottom', size = 12)

```

```python
plt.figure(figsize=(10,5))

max_temp = plt.plot(all_mean_temp['월'],all_mean_temp['최대'],marker='.', linestyle='-', label='최대')
for xi, yi in zip(all_mean_temp['월'], all_mean_temp['최대']):
    plt.annotate(f'{yi:.1f}', (xi, yi), textcoords="offset points", xytext=(0,5), ha='center')

min_temp = plt.plot(all_mean_temp['월'],all_mean_temp['최소'],marker='.', linestyle='-', label='최소')
for xi, yi in zip(all_mean_temp['월'], all_mean_temp['최소']):
    plt.annotate(f'{yi:.1f}', (xi, yi), textcoords="offset points", xytext=(0,5), ha='center')

plt.xticks(all_mean_temp['월'])
plt.title('월 최대, 최소 기온(전국 10년 평균)')
plt.xlabel('월')
plt.ylabel('기온')
plt.legend(loc='upper right')
plt.ylim([-5, 30]) 
plt.show()
```
